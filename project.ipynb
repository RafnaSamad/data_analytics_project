# U.S. Heart Disease Mortality Analysis

### INTRODUCTION

This dataset contains county- and state-level information on heart disease mortality among U.S. adults aged 35 and above for the years 2019â€“2021. 
It provides mortality rates and related indicators across different geographic regions, making it suitable for analyzing public health trends,
regional disparities, and risk patterns.

### KEY ATTRIBUTES

YEAR                : Year in which the data was collected.
LocationAbbr        : Abbreviation code of the location (state or territory code)
LocationDesc        : Full name of the geographic location (state, territory, or country).
GeographicLevel     : Level of geographic detail (State, County, or Territory).
DataSource          : Source organization that provided the data.
Class               : Broad classification of the health indicator.
Topic               : Main health topic covered by the data (e.g., Heart Disease).
Data_Value          : Numeric value representing the measured health indicator.
Data_Value_Unit     : Unit of measurement for the data value (e.g., rate per 100,000).
Data_Value_Type	    :Type of data value (e.g., Age-adjusted rate, crude rate).
Data_Value_Footnote :	Additional notes or explanations related to the data value.
StratificationCategory1: First category used to stratify the data (e.g., Age Group).
Stratification1	    : Specific group within the first stratification category.
StratificationCategory2: Second category used for stratification (if applicable).
Stratification2	    : Specific group within the second stratification category.
TopicID	            :Unique identifier for the health topic.
LocationID          : Unique numeric identifier for the geographic location.
Y_lat               : Latitude coordinate of the geographic location.
X_lon               : Longitude coordinate of the geographic location.
Georeference        : Geographic reference combining latitude and longitude for mapping purposes.

```python
g = df[(df['GeographicLevel'] == 'State') &(df['Stratification1'].isin(['Male', 'Female'])) &(df['Data_Value'].notna())].copy()
pt = g.pivot_table(
    values='Data_Value',
    index='LocationDesc',
    columns='Stratification1',
    aggfunc='mean'
)
pt = pt.assign(Overall=pt.mean(axis=1)).sort_values('Overall', ascending=False).drop(columns='Overall').head(10)
plt.figure(figsize=(6, 10),facecolor="skyblue")
plt.imshow(pt.values, aspect='auto')
plt.xticks(range(len(pt.columns)), pt.columns)
plt.yticks(range(len(pt.index)), pt.index)
plt.xlabel("Gender")
plt.ylabel("State")
plt.title("Heart Disease Mortality by State and Gender (Top 10 States)")
plt.colorbar(label="Mortality Rate")
plt.show()
```

```python
df['X_lon'] = pd.to_numeric(df['X_lon'], errors='coerce')
df['Y_lat'] = pd.to_numeric(df['Y_lat'], errors='coerce')
state_overall = df[
    (df['GeographicLevel'] == 'State') &
    (df['Stratification1'] == 'Overall') &
    df['Data_Value'].notna() &
    df['X_lon'].notna() &
    df['Y_lat'].notna()
].copy()
state_summary = (state_overall.groupby(['LocationDesc','X_lon','Y_lat'], as_index=False)['Data_Value'].mean().sort_values('Data_Value', ascending=False))
top10 = state_summary.head(10)
plt.figure(figsize=(10,6),facecolor="skyblue")
sc = plt.scatter(top10['X_lon'], top10['Y_lat'], c=top10['Data_Value'], s=160)
for i in range(len(top10)):
    plt.text(top10['X_lon'].iloc[i], top10['Y_lat'].iloc[i], top10['LocationDesc'].iloc[i], fontsize=8)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Top 10 U.S. States by Heart Disease Mortality Rate (Overall)")
plt.colorbar(sc, label="Mortality Rate")
plt.show()
```

```python
combo_df = df[
    (df['GeographicLevel'] == 'Nation') &
    (df['Stratification1'].isin(['Male', 'Female'])) &
    df['Stratification2'].notna() &
    df['Data_Value'].notna()
]
pivot = combo_df.pivot_table(
    values='Data_Value',
    index='Stratification2',
    columns='Stratification1',
    aggfunc='mean'
)
x = np.arange(len(pivot.index))
width = 0.35
plt.figure(figsize=(12,6))
plt.bar(x - width/2, pivot['Male'], width, label='Male')
plt.bar(x + width/2, pivot['Female'], width, label='Female')
plt.xlabel("Race / Ethnicity (Stratification 2)")
plt.ylabel("Average Heart Disease Mortality Rate")
plt.title("Grouped Bar Chart: Heart Disease Mortality by Gender and Race")
plt.xticks(x, pivot.index, rotation=45, ha='right')
plt.legend()
plt.show()
```

```python
black_state_df = df[(df['GeographicLevel'] == 'State') &(df['Stratification2'] == 'Black') &(df['Stratification1'] == 'Overall') &
    df['Data_Value'].notna() &
    df['X_lon'].notna() &
    df['Y_lat'].notna()
].copy()
black_state_summary = (
    black_state_df
    .groupby(['LocationDesc', 'X_lon', 'Y_lat'], as_index=False)['Data_Value']
    .mean()
    .sort_values('Data_Value', ascending=False)
)
top20_black_states = black_state_summary.head(20)
plt.figure(figsize=(11, 7))
scatter = plt.scatter(
    top20_black_states['X_lon'],
    top20_black_states['Y_lat'],
    c=top20_black_states['Data_Value'],
    s=180
)
for i in range(len(top20_black_states)):
    plt.text(
        top20_black_states['X_lon'].iloc[i],
        top20_black_states['Y_lat'].iloc[i],
        top20_black_states['LocationDesc'].iloc[i],
        fontsize=8
    )
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Top 20 U.S. States by Heart Disease Mortality (Black Population)")
plt.colorbar(scatter, label="Mortality Rate")
plt.show()
```

### INTRODUCTION

Heart disease remains one of the leading causes of mortality in the United States. Understanding how mortality varies across geographic regions and demographic groups is essential for effective public health planning. This project applies exploratory data analysis techniques to investigate patterns in heart disease mortality by state, gender, and race/ethnicity among U.S. adults aged 35 and above.


### METHODOLOGY

The dataset was cleaned and prepared by converting mortality values to numeric format and removing missing or suppressed values where necessary. Data was filtered by geographic level (nation, state, and county) and stratified by gender (Stratification 1) and race/ethnicity (Stratification 2). Aggregation was performed using group-by operations, and insights were derived through visualizations including bar charts, line charts, boxplots, violin plots, slope charts, heatmaps, and geographic maps.

### KEY INSIGHTS

The analysis demonstrates substantial disparities in heart disease mortality across states and demographic groups. Males consistently experience higher mortality rates than females. Racial analysis reveals that the Black population faces disproportionately higher mortality, particularly in specific states. Geographic visualizations further highlight regional clustering of high mortality rates, suggesting location-based health inequalities.


### CONCLUTION

The findings of this study indicate that heart disease mortality among U.S. adults aged 35 and above is strongly influenced by geographic location, gender, and race/ethnicity. The results emphasize the importance of targeted public health strategies aimed at high-risk populations and regions. Addressing disparities in healthcare access and preventive services is critical to reducing heart disease mortality.


### RECOMMENDATIONS

Targeted interventions should focus on high-risk states and populations, particularly males and the Black population. Improving preventive care, healthcare access, and awareness can help reduce heart disease mortality.
